Extraemos unos cuantos frames clave

```python
import cv2
from PIL import Image
import numpy as np

def downsample_video(video_path: str, num_frames: int = 10) -> list:
    """
    Downsample the video by extracting a fixed number of frames evenly spaced throughout the video.

    Args:
        video_path: Path to the video file.
        num_frames: The number of frames to extract.

    Returns:
        A list of tuples, each containing a PIL image and its corresponding timestamp.
    """
    vidcap = cv2.VideoCapture(video_path)
    total_frames = int(vidcap.get(cv2.CAP_PROP_FRAME_COUNT))
    fps = vidcap.get(cv2.CAP_PROP_FPS)

    if total_frames <= 0 or fps <= 0:
        raise ValueError("Could not determine video properties (frame count or FPS).")

    frame_indices = np.linspace(0, total_frames - 1, num_frames, dtype=int)
    frames = []

    for i in frame_indices:
        vidcap.set(cv2.CAP_PROP_POS_FRAMES, i)
        success, image = vidcap.read()
        if not success:
            print(f"Warning: Unable to read frame at index {i}.")
            continue
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB
        pil_image = Image.fromarray(image)
        timestamp = round(i / fps, 2)
        frames.append((pil_image, timestamp))

    vidcap.release()
    return frames
```

```python
frames = downsample_video("cats_1.mp4")
```

Creamos una estructura de prompt al que le añadiremos las imágenes

```python
messages = [
    {
        "role": "system",
        "content": [{"type": "text", "text": "You are a helpful assistant."}]
    },

    {
        "role": "user",
        "content": [
            {"type": "text", "text": f"What is happening in this video? Summarize the events."}]
    }
]
```

Añadimos las imágenes

```python
for frame in frames:
    image, timestamp = frame
    messages[1]["content"].append({"type": "text", "text": f"Frame {timestamp}:"})
    image.save(f"image_{timestamp}.png")
    messages[1]["content"].append({"type": "image", "url": f"image_{timestamp}.png"})
```

```python
messages
```

Tokenizamos nuestro prompt, inferimos e imprimimos la respuesta.

```python
inputs = processor.apply_chat_template(
    messages, add_generation_prompt=True, tokenize=True,
    return_dict=True, return_tensors="pt"
).to(model.device)

input_len = inputs["input_ids"].shape[-1]

generation = model.generate(**inputs, max_new_tokens=100, do_sample=False)
generation = generation[0][input_len:]

decoded = processor.decode(generation, skip_special_tokens=True)
print(decoded)
```
